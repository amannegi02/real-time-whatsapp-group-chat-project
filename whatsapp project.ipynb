```python
pattern='\d{2}/\d{2}/\d{2},\s\d{1,2}:\d{2}\s[a,p]{1}m\s-\s'
```

```python
messages=re.split(pattern,data)
```

```python
messages=messages[1:]
```

```python
dates=re.findall(pattern,data)
```

```python
len(dates)
```

```python
dates
```

```python
date=list(map(lambda x:x.replace("\u202f",""),dates))
```

```python
len(date)
```

```python
date
```

```python
df=pd.DataFrame({"user_message":messages,"message_date":date})
df
```

```python
df.head(2)
```

```python
df.info()
```

```python
df["message_date"]=pd.to_datetime(df["message_date"],format='%d/%m/%y, %I:%M%p - ')  #%p for am,pm


```

```python
df.dtypes
```

```python
df["year"]=df["message_date"].dt.year
df["month"]=df["message_date"].dt.month_name()
df["day"]=df["message_date"].dt.day
df["hour"]=df["message_date"].dt.hour
df["month"]=df["message_date"].dt.month
df["day_name"]=df["message_date"].dt.day_name()
df["Date"]=df["message_date"].dt.date
df["minute"]=df["message_date"].dt.minute
df.sample(6)
```

```python
users_1=[]
messages_1=[]
for i in df["user_message"]:
    if (("created group" in i) | ("changed the group" in i)):
        li=[]
        li.append(i)
        #print(li)
    elif (("created group" not in i) | ("changed the group" not in i)) :
        if ":" in i:
            li=i.split(":",maxsplit=1)
            #print(li)
        else:
            li=[]
            li.append(i)
    if len(li)>1:
        users_1.append(li[0])
        messages_1.append(li[1])
    else:
        users_1.append("group_notification")
        messages_1.append(li[0])

```

```python
print(len(users_1))
```

```python
print(len(messages_1))
```

```python
df["user_name"]=users_1
df["message"]=messages_1
```
